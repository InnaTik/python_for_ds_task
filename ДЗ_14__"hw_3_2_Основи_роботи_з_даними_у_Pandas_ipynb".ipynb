7.1. Середнє значення валового доходу більше в місті Mandalay.

7.2. Гіпотези

1. Вплив типу клієнта

    Чому: Постійні клієнти можуть робити більші або частіші покупки.

    Правдивість гіпотези: Порівняння середніх значень gross income по типу клієнта (Customer type) у розрізі міст (City).

    Висновки для прийняття ріщення: Нові програми лояльності для утримання поточних клієнтів та заохочення нових.


```python
# Вплив наявного членства для міста Yangon
filter1 = (df['City'] == 'Yangon') #| (df['City'] == 'Mandalay')
d1 = df[filter1].groupby('Customer type')['gross income'].mean()
```

```python
# Вплив наявного членства для міста Mandalay
filter2 = (df['City'] == 'Mandalay')
d2 = df[filter2].groupby('Customer type')['gross income'].mean()
```

```python
combined = pd.concat([d1, d2],axis=1)
combined.columns = ['Yangon', 'Mandalay']
combined
```

2. Розрахунок певним методом оплати

    Чому: Клієнти можуть робити більший чек при оплаті кредитною карткою або електронним гаманцем.

    Правдивість гіпотези: Порівняння середніх значень gross income по типу оплати (Payment) у розрізі міст (City). Також можна додати розріз по типу клієнта (Customer type), бо є вірогідність, якщо клієнт Member, то і безготівкові оплати більші.

    Висновки для прийняття ріщення: Додаткові умови програми лояльності для користувачів кридитних карток (кешбек за оплату карткою). Розробка нових методів для оплати товару (як приклад Fast Line в METRO).


```python
# Вплив методу оплати для міста Yangon
filter1_1 = (df['City'] == 'Yangon')
d1_1 = df[filter1_1].groupby('Payment')['gross income'].mean()
```

```python
# Вплив наявного членства для міста Mandalay
filter2_1 = (df['City'] == 'Mandalay')
d2_1 = df[filter2_1].groupby('Payment')['gross income'].mean()
```

```python
combined_1 = pd.concat([d1_1, d2_1],axis=1,)
combined_1.columns = ['Yangon', 'Mandalay']
combined_1
```

3. У місті з вищим gross income покупці частіше обирають дорожчі товари

    Чому: Вищий середній дохід (gross income) може бути пов'язаний із вищими цінами (Unit price) на групу товарів (Product line) .

    Правдивість гіпотези: Порівняти середнє значення (Unit price) у містах (City). Проаналізувати, які (Product line) купують частіше в місті з вищим доходом.

    Висновки для прийняття ріщення: Оптимізація асортимента та введення нових маркетингових активностей.


```python
# Порівняння у розрізі категорій по містам: середня ціна, частота покупки, середній валовий дохід
filter1_2 = (df['City'] == 'Yangon')
d1_2 = df[filter1_2].groupby('Product line')['Unit price'].mean()

filter2_2 = (df['City'] == 'Mandalay')
d2_2 = df[filter2_2].groupby('Product line')['Unit price'].mean()

filter1_3 = (df['City'] == 'Yangon')
d1_3 = df[filter1_3]['Product line'].value_counts()

filter2_3 = (df['City'] == 'Mandalay')
d2_3 = df[filter2_3]['Product line'].value_counts()

filter1_4 = (df['City'] == 'Yangon')
a = df[filter1_4].groupby('Product line')['gross income'].mean()

filter2_4 = (df['City'] == 'Mandalay')
b = df[filter2_4].groupby('Product line')['gross income'].mean()


combined_3 = pd.concat([d1_2, d1_3, a, d2_2, d2_3, b],axis=1,)
combined_3.columns = ['Yangon mean price', 'Yangon count', 'Yangon mean gross income', 'Mandalay mean price', 'Mandalay count', 'Mandalay mean gross income']
combined_3
```
