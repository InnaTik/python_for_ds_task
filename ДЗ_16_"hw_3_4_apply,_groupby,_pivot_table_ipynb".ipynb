2. Чи має вплив на загальний дохід (Total) рейтинг задоволення клієнтів (Rating) тої чи іншої категорії товарів (Product line).

    Чому: Категорії товарів з нижчими рейтингами (Rating) можуть мати нижчий дохід (Total) через незадоволення клієнтів, як і товари з високим рейтингом (Rating) мати весокий загальних дохід (Total).

    Перевірка гіпотези: Порівняння середнього по рейтингу (Rating) та загального доходу (Total) в розрізі категорій (Product line).
  
    Висновки: Загалом, немає чіткої залежності між рейтингом (Rating) і загальним доходом (Total). Ймовірно, на дохід впливає не тільки задоволеність, а й необхідність товару, акційні пропозиції, розташування та інше.

    Найгірший середній рейтинг у "Home and lifestyle" - 6.84, але загальний дохід не найменший, як і середній валовий дохід. Товари купують, бо є потреба.
    "Food and beverages" має найвищий середній рейтинг - 7.11, та найвищий дохід - 56 144,84. Що підтверджує гіпотезу, що чим більш задоволений клієнт, тим і більший дохід.

    Примітки: Навіть з низьким рейтингом окремі категорії мають хороші продажі - це може бути сигналом до покращення якості обслуговування. Також можна робити більше маркетингових активностей в категоріях де "рейтинг = вищий дохід".

```python
df['gross income sum'] =df['gross income']
df.groupby('Product line').agg({'Rating': 'mean', 'Total': 'sum', 'Invoice ID':'count', 'gross income':'mean', 'gross income sum':'sum'}).sort_values(by='Rating', ascending=False).round(2)
```

3. Загальний дохід (Total), який приносить кожна категорія товарів (Product line) в залежності від статі клієнта (Gender).

    Чому: Категорія товара може бути популярна серед певної статі, що може впливати на дохід через певну категорію товара.

    Перевірка гіпотези: Побудова зведеної таблиці де буде загальний дохід (Total) та його відсоткове співвідношення згідно статі клієнта (Gender) по категорії товара (Product line).

    Висновки: Присутня частина категорій товару, "Electronic accessories" та "Sports and travel",  по котрим продажі майже рівні, тобто стать клієнта не важлива.

    Є категорії, де дохід переважає зі сторони певної стані:
      *   Жінки: "Fashion accessories" (56,05%), "Food  and beveranges" (59,08%), "Home and lifestyle" (55,77%);
      *   Чоловіки: "Health and beauty" (62,27%).
    
    Тобто стать дійсно впливає на дохід у певних категоріях товару.

    Примітки: Згідно висновків, можна робити додаткові запуски реклами з акційними пропозиціями в соціальних мережах та середовищі, де жінка проводе більшість часу.

```python
pivot = pd.pivot_table(df, values='Total', index='Gender', columns='Product line', aggfunc='sum', fill_value=0)

col_totals = pivot.sum(axis=0)

percent_pivot = (pivot / col_totals) * 100

new_table = percent_pivot.round(2).T

pivot1 = pd.pivot_table(df, values='Total', index='Product line', columns='Gender', aggfunc='sum', fill_value=0)

result = pd.concat([pivot1 , new_table], axis=1)
result

```
