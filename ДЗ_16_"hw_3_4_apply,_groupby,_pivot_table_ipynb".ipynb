2. –ß–∏ –º–∞—î –≤–ø–ª–∏–≤ –Ω–∞ –∑–∞–≥–∞–ª—å–Ω–∏–π –¥–æ—Ö—ñ–¥ (Total) —Ä–µ–π—Ç–∏–Ω–≥ –∑–∞–¥–æ–≤–æ–ª–µ–Ω–Ω—è –∫–ª—ñ—î–Ω—Ç—ñ–≤ (Rating) —Ç–æ—ó —á–∏ —ñ–Ω—à–æ—ó –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤ (Product line).

    –ß–æ–º—É: –ö–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤ –∑ –Ω–∏–∂—á–∏–º–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞–º–∏ (Rating) –º–æ–∂—É—Ç—å –º–∞—Ç–∏ –Ω–∏–∂—á–∏–π –¥–æ—Ö—ñ–¥ (Total) —á–µ—Ä–µ–∑ –Ω–µ–∑–∞–¥–æ–≤–æ–ª–µ–Ω–Ω—è –∫–ª—ñ—î–Ω—Ç—ñ–≤, —è–∫ —ñ —Ç–æ–≤–∞—Ä–∏ –∑ –≤–∏—Å–æ–∫–∏–º —Ä–µ–π—Ç–∏–Ω–≥–æ–º (Rating) –º–∞—Ç–∏ –≤–µ—Å–æ–∫–∏–π –∑–∞–≥–∞–ª—å–Ω–∏—Ö –¥–æ—Ö—ñ–¥ (Total).

    –í–∏—Å–Ω–æ–≤–∫–∏: –ó–∞–≥–∞–ª–æ–º, –Ω–µ–º–∞—î —á—ñ—Ç–∫–æ—ó –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –º—ñ–∂ —Ä–µ–π—Ç–∏–Ω–≥–æ–º (Rating) —ñ –∑–∞–≥–∞–ª—å–Ω–∏–º –¥–æ—Ö–æ–¥–æ–º (Total). –ô–º–æ–≤—ñ—Ä–Ω–æ, –Ω–∞ –¥–æ—Ö—ñ–¥ –≤–ø–ª–∏–≤–∞—î –Ω–µ —Ç—ñ–ª—å–∫–∏ –∑–∞–¥–æ–≤–æ–ª–µ–Ω—ñ—Å—Ç—å, –∞ –π –Ω–µ–æ–±—Ö—ñ–¥–Ω—ñ—Å—Ç—å —Ç–æ–≤–∞—Ä—É, –∞–∫—Ü—ñ–π–Ω—ñ –ø—Ä–æ–ø–æ–∑–∏—Ü—ñ—ó, —Ä–æ–∑—Ç–∞—à—É–≤–∞–Ω–Ω—è —Ç–∞ —ñ–Ω—à–µ.

    –ù–∞–π–≥—ñ—Ä—à–∏–π —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä–µ–π—Ç–∏–Ω–≥ —É "Home and lifestyle" - 6.84, –∞–ª–µ –∑–∞–≥–∞–ª—å–Ω–∏–π –¥–æ—Ö—ñ–¥ –Ω–µ –Ω–∞–π–º–µ–Ω—à–∏–π, —è–∫ —ñ —Å–µ—Ä–µ–¥–Ω—ñ–π –≤–∞–ª–æ–≤–∏–π –¥–æ—Ö—ñ–¥. –¢–æ–≤–∞—Ä–∏ –∫—É–ø—É—é—Ç—å, –±–æ —î –ø–æ—Ç—Ä–µ–±–∞.
    "Food and beverages" –º–∞—î –Ω–∞–π–≤–∏—â–∏–π —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä–µ–π—Ç–∏–Ω–≥ - 7.11, —Ç–∞ –Ω–∞–π–≤–∏—â–∏–π –¥–æ—Ö—ñ–¥ - 56 144,84. –©–æ –ø—ñ–¥—Ç–≤–µ—Ä–¥–∂—É—î –≥—ñ–ø–æ—Ç–µ–∑—É, —â–æ —á–∏–º –±—ñ–ª—å—à –∑–∞–¥–æ–≤–æ–ª–µ–Ω–∏–π –∫–ª—ñ—î–Ω—Ç, —Ç–∏–º —ñ –±—ñ–ª—å—à–∏–π –¥–æ—Ö—ñ–¥.

    –ü—Ä–∏–º—ñ—Ç–∫–∏: –ù–∞–≤—ñ—Ç—å –∑ –Ω–∏–∑—å–∫–∏–º —Ä–µ–π—Ç–∏–Ω–≥–æ–º –æ–∫—Ä–µ–º—ñ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó –º–∞—é—Ç—å —Ö–æ—Ä–æ—à—ñ –ø—Ä–æ–¥–∞–∂—ñ - —Ü–µ –º–æ–∂–µ –±—É—Ç–∏ —Å–∏–≥–Ω–∞–ª–æ–º –¥–æ –ø–æ–∫—Ä–∞—â–µ–Ω–Ω—è —è–∫–æ—Å—Ç—ñ –æ–±—Å–ª—É–≥–æ–≤—É–≤–∞–Ω–Ω—è. –¢–∞–∫–æ–∂ –º–æ–∂–Ω–∞ —Ä–æ–±–∏—Ç–∏ –±—ñ–ª—å—à–µ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤–∏—Ö –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–µ–π –≤ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö –¥–µ "—Ä–µ–π—Ç–∏–Ω–≥ = –≤–∏—â–∏–π –¥–æ—Ö—ñ–¥".

    –ß–æ–º—É: –ö–∞—Ç–µ–≥–æ—Ä—ñ—è —Ç–æ–≤–∞—Ä–∞ –º–æ–∂–µ –±—É—Ç–∏ –ø–æ–ø—É–ª—è—Ä–Ω–∞ —Å–µ—Ä–µ–¥ –ø–µ–≤–Ω–æ—ó —Å—Ç–∞—Ç—ñ, —â–æ –º–æ–∂–µ –≤–ø–ª–∏–≤–∞—Ç–∏ –Ω–∞ –¥–æ—Ö—ñ–¥ —á–µ—Ä–µ–∑ –ø–µ–≤–Ω—É –∫–∞—Ç–µ–≥–æ—Ä—ñ—é —Ç–æ–≤–∞—Ä–∞.

    –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –≥—ñ–ø–æ—Ç–µ–∑–∏: –ü–æ–±—É–¥–æ–≤–∞ –∑–≤–µ–¥–µ–Ω–æ—ó —Ç–∞–±–ª–∏—Ü—ñ –¥–µ –±—É–¥–µ –∑–∞–≥–∞–ª—å–Ω–∏–π –¥–æ—Ö—ñ–¥ (Total) —Ç–∞ –π–æ–≥–æ –≤—ñ–¥—Å–æ—Ç–∫–æ–≤–µ —Å–ø—ñ–≤–≤—ñ–¥–Ω–æ—à–µ–Ω–Ω—è –∑–≥—ñ–¥–Ω–æ —Å—Ç–∞—Ç—ñ –∫–ª—ñ—î–Ω—Ç–∞ (Gender) –ø–æ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä–∞ (Product line).

    –Ñ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó, –¥–µ –¥–æ—Ö—ñ–¥ –ø–µ—Ä–µ–≤–∞–∂–∞—î –∑—ñ —Å—Ç–æ—Ä–æ–Ω–∏ –ø–µ–≤–Ω–æ—ó —Å—Ç–∞–Ω—ñ:
      *   –ñ—ñ–Ω–∫–∏: "Fashion accessories" (56,05%), "Food  and beveranges" (59,08%), "Home and lifestyle" (55,77%);
